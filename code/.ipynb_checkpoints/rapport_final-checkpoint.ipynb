# Aperçus des Datamarts: 2ème partie

## Schémas 2 et 3: Restaurants à Paris ou Avignon ayant subit un contrôle sanitaire en 2015

Nos deux prochains cubes ont pour base le fichier csv, accessible depuis le lien suivant:
https://www.data.gouv.fr/fr/datasets/experimentation-de-la-mise-en-transparence-des-resultats-des-controles-officiels-en-restauration-commerciale-a-paris-et-avignon

Celui-ci recense toutes les contrôles sanitaires officiels réalisés durant l'année 2015. Les données rendues accessibles par la mise en transparence des résultats des contrôles officiels en hygiène alimentaire des restaurants.

Légende des Évaluations des restaurants:

* 30: Niveau d’hygiène « à améliorer ; mesures correctives requises » : il s’agit des établissements pour lesquels les déficiences constatées nécessitent la mise en place de mesures spécifiques (nettoyage, travaux, formation, …), mis en demeure de procéder à des mesures correctives, dont l’efficacité sera vérifiée lors d’un nouveau contrôle.


* 20: Niveau d’hygiène « acceptable » : établissements présentant des anomalies qui ne justifient pas l’application de mesures obligatoires par l’administration.


* 10: Niveau d’hygiène « bon » : établissements conformes à la réglementation ou présentant quelques anomalies mineures.


### Preprocessing:

Le fichier récupéré du site ci-dessus nous informe du nom, de la raison sociale et de son numéro de SIRET associé, du lieu (adresse et géoposition), de la date et de l'évaluation associée. 

Nous avons choisit de combiner ces informations avec d'autres, dans l'optique de faire une analyse par rapport à l'évaluation et la côte vis-vis des consomateurs du restaurant ayant subi un contrôle. A l'aide de python et de BeautifulSoup, nous avons pu créer deux webscrappers:

* un qui, à partir de l'adresse et de l'enseigne du restaurant, se sert du moteur de recherche de yelp.fr afin de récupérer le restaurant, ainsi que les reviews et les ratings faits par les internautes. Des informations sur les reviewers sont également récupérés.

* et l'autre, qui, à partir du numéro de SIRET de l'entreprise possédant le restaurant, extrait les informations disponibles associées à l'entreprise: le capital, le type d'entreprise, etc...


### aperçu fichier.csv

## I - Représentation des données


### 1 - Représentation vectorielle
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Nous disposons dans notre dataset des données hétérogènes: des variables quantitatives (i.e. les ratings, qui sont des réels), mais aussi des variables qualitatives (i.e.  l'arrondissement, est une variable qualitative comme on considère qu'il n'existe pas de 
lien pertinent entre les différentes valeurs qu'elle peut prendre; il n'y a pas d'échelle entre les différents arrondissements).


&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Certains modèles d'apprentissage, comme par exemple les arbres de décision gèrent à la fois des données qualitatives et quantitatives. Mais d'autres ne prennent uniquement des données quantitatives (numériques) pour entrée: il faut donc trouver un moyen de transformer des données qualitatives en données quantitatives s'il l'on veut utiliser ces modèles. Il existe une méthode qui consiste à représenter les données qualitatives sous forme de vecteurs. Cette méthode est analogue à la représentation d'un document textuel sous la forme de Bag-of-Words:si l'on veut représenter une variable qualitative $x$ qui tire ses valeurs d'un espace discret de cardinalité $n$, on peut construire un vecteur de même taille, où l'on mets à 1 son composant $i$ lorsque la variable associée est tirée, et 0 sinon. De nombreuses extensions de cette représentation vectorielle existent: nous pouvons également compter le nombre de fois où la variable tire une certaine valeur.

### 1 - Prédiction de l'évalution sanitaire
Les meilleurs résultats ont étés obtenus avec une machine à vecteurs support, à l'aide d'un noyau anova, défini comme ci dessous:

<img src="anova.png">


&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**accuracy: 57,62%**
<img src="res_svm.png">

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Les résultats obtenus ne sont visiblement par très bons, mais restent néeanmoins supérieurs à l'aléatoire, qui est de 33,33%. La répartition des restaurants par rapport aux classes est très inégale: il y a très peu de restaurants (10) qui ont une évalution de 30. Nous avons néanmoins souhaités garder cette distribution par rapport aux classes car elle pourrait être représentative de la répartition des évaluations pour d'autres restaurants, même si il peut y avoir un biais car l'établissement public aurait très bien pu effectuer des enquêtes pour tenter de déterminer les restaurants qui ne respectent pas les normes d'hygiène. Nous avons cependant très peu de données, il est donc difficile pour notre modèle de capter l'information discriminante.


### 2 - Prédiction des ratings

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Les ratings sont des entiers compris entre 1 et 5. Comme la signification du rating de 3 varie selon l'internaute, nous avons supprimé tout les lignes qui possédaient un rating de 3. Nous avons ensuite rassemblés les ratings de 1 et de 2, et ceux de 4 et de 5, pour constituer des classes de ratings négatifs et positifs (respectivement). Nous avons donc transformés notre problème en un problème de classification binaire. Voici la matrice de confusion des résultats. Il peut être interessant de noter que les deux classes n'ont pas la même quantité d'exemples; nous avons souhaités garder cette inégalité de répartition car nous supposons que cette distribution est bien représentative de la répartition des ratings pour d'autres restaurants.


&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**accuracy: 90,46%**
